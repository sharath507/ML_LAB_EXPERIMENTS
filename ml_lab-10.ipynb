```python
# prompt: Clustering Algorithms
# Objective: Implement various clustering algorithms.
# Tasks: Apply K-means, Spectral, and Hierarchical clustering on a dataset and
# compare the clustering results.

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans, SpectralClustering, AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load your dataset (replace 'your_dataset.csv' with your file)
data = pd.read_csv('/content/mango2.csv')

# Select features for clustering (replace with your feature columns)
# features = data[['feature1', 'feature2', 'feature3']]
features = data
# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# K-means clustering
kmeans = KMeans(n_clusters=2, random_state=0)  # Adjust n_clusters as needed
kmeans_labels = kmeans.fit_predict(scaled_features)

# Spectral clustering
spectral = SpectralClustering(n_clusters=2, assign_labels='kmeans', random_state=0) # Adjust n_clusters as needed
spectral_labels = spectral.fit_predict(scaled_features)

# Hierarchical clustering
hierarchical = AgglomerativeClustering(n_clusters=2)  # Adjust n_clusters as needed
hierarchical_labels = hierarchical.fit_predict(scaled_features)

# Evaluate clustering performance (Silhouette Score)
kmeans_silhouette = silhouette_score(scaled_features, kmeans_labels)
spectral_silhouette = silhouette_score(scaled_features, spectral_labels)
hierarchical_silhouette = silhouette_score(scaled_features, hierarchical_labels)

print(f"K-means Silhouette Score: {kmeans_silhouette}")
print(f"Spectral Silhouette Score: {spectral_silhouette}")
print(f"Hierarchical Silhouette Score: {hierarchical_silhouette}")

# Visualize the clustering results (example for 2D data)
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=kmeans_labels)
plt.title("K-means Clustering")

plt.subplot(1, 3, 2)
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=spectral_labels)
plt.title("Spectral Clustering")

plt.subplot(1, 3, 3)
plt.scatter(scaled_features[:, 0], scaled_features[:, 1], c=hierarchical_labels)
plt.title("Hierarchical Clustering")

plt.show()
```
